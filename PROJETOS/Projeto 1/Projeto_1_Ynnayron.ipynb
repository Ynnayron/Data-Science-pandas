**(b) qual a proporção de pessoas ilesas e de feridos graves por mês em cada um dos anos
considerados? (proporção = numero de ilesos ou feridos graves/ número total de
pessoas envolvidas no acidente);**

```python
import pandas as pd
import datetime
import numpy as np

#2018

#Feridos Graves
fr_ano_2018=pd.DataFrame(dados2018[['data_inversa','feridos_graves','pessoas']])
fr_ano_2018['data_inversa']=pd.to_datetime(fr_ano_2018['data_inversa']) 
fr_ano_2018['Meses']=fr_ano_2018['data_inversa'].dt.month
fr_ano_2018=fr_ano_2018.groupby('Meses')
fr_ano_2018=fr_ano_2018.sum()
fr_ano_2018['GRAVES']=fr_ano_2018['feridos_graves']/fr_ano_2018['pessoas']
graves2018=fr_ano_2018['GRAVES']

#ilesos
ano_2018=pd.DataFrame(dados2018[['data_inversa','ilesos','pessoas']])
ano_2018['data_inversa']=pd.to_datetime(ano_2018['data_inversa']) 
ano_2018['Meses']=ano_2018['data_inversa'].dt.month
ano_2018=ano_2018.groupby('Meses')
ano_2018=ano_2018.sum()
ano_2018['ILESOS']=ano_2018['ilesos']/ano_2018['pessoas']
ilesos2018=ano_2018['ILESOS']
df_2018=pd.concat([graves2018,ilesos2018],	axis=1)
dataframe_2018=pd.DataFrame(df_2018)
print("###########################")
print("           2018           ")
print("###########################")
dataframe_2018
```
